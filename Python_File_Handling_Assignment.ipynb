# Theory Questions
1. What is the difference between interpreted and compiled languages?
Interpreted languages execute the source code line by line at runtime, while compiled languages first convert the entire source code into machine code before execution.
Example: Python is an interpreted language. C and C++ are compiled languages. Interpreted languages are easier to debug but slower in execution.

2. What is exception handling in Python?
Exception handling is a mechanism in Python to handle runtime errors so that the program does not crash suddenly. It uses try, except, else, and finally blocks.

3. What is the purpose of finally block in exception handling?
The finally block contains code that will always execute, whether an exception occurs or not. It mainly used for cleanup actions like closing files or releasing resources.
Example: CLosing a file after reading or writing.

4. What is logging in Python?
Logging is the process of recording messages about the program's execution. It helps developers track errors, warnings, and important events during runtime. Python provides the logging module for this purpose.

5. What is the significance of the __del__ method in Python?
The __del__ method is a destructor. It is automatically called when an object is about to be destroyed(garbage collected). It is used to perform cleanup tasks like closing database connections.

6. What is the difference between import and from ... import in Python?
- import module imports the entire module. We access functions using module.function().
- from module import function imnport only specific functions or classes from the module. We can use them directly without the module name.

7. How can you handle multiple exceptions in Python?
We can handle multiple exceptions by using a tuple in the except block or by writing multiple except blocks.
```
try:
  ...
except (ZeroDivisionError, ValueError) as e:
    print("Error occured:", e)
```

8. What is the purpose of the with statement when handling files in Python?
The with statement is used for context management. It automatically closes the file after the block is exectuted, even if an error occurs. It makes file handling safer and cleaner.

9. What is the difference between multithreading and multiprocessing?
- Multithreading: Multiple threads run within the same process and share memory. It is good for I/O-bound tasks.
- Multiprocessing: Multiple processes run independently with their own memory space. It is useful for CPU-bound tasks and bypasses GIL.

10. What are the advantages of using logging in a program?
- Helps in debugging and troubleshooting
- Records important events and errors
- Supports different log levels(DEBUG, INFO, WARNING, ERROR, CRITICAL)
- Logs can be saved to files with rotation
- Makes maintenance easier in large applications

11. What is memory management in Python?
Memory management is the process of allocating and deallocating memory for objects. Python uses a private heap space and automatically manages memory using reference counting and garbage collection.

12. What are the basic steps involved in exception handling in Python?
 1. Write risky code inside try block
 2. Handle exceptions in except block
 3.(Optional) Use else block if no exception occurs
 4. Use finally block for cleanup code

13.Why is memory management important in Python?
Good memory management prevents leak, reduce crashes, improves performance, and ensures efficient use of system resources, especially in long-running applications.

14. What is the role of try and except in exception handling?
 - try block contains the code that might raise an exception.
 - except block catches and handles the exception so the program can continue running gracefully.

15. How does Python's garbage collection system work?
Python uses reference counting(counts how many references point to an object) and a cyclic garbage collector to detect and remove circular references. When reference count becomes zero, the object is automatically deleted.

16. What is the purpose of the else block in exception handling?
The else block executes only if no exception occurs in the try block. It is used to run code that should execute only when the try block succeeds.

17. What are the common logging levels in Python?
The common logging levels are:
 - DEBUG(lowest)
 - INFO
 - WARNING
 - ERROR
 - CRITICAL(highest)

18. What is the difference between os.fork() and multiprocessing in Python?
os.fork() is a low-level Unix function that creates a child process by duplicating the current process. multiprocessing is a high-level Python module that provides easier and cross-platform process creation with better abstractions.

19. What is the importance of closing a file in Python?
Closing a file releases system resources, flushes any buffered data to the disk, and prevents file corruption or resources leaks.

20. What is the difference between file.read() and file.readline() in Python?
 - read() reads the entire file content at once and returns it as a string.
 - readline() reads only one line at a time and returns it as a string.

21. What is the logging module in Python used for?
 The logging module is used to record log messages at different severity levels. It is more flexible and powwerful than using print() statements.

22. What is the os module in Python used for in file handling?
The os module provides functions to interact with the operating system, such as creating, deleting, renaming files and directories, checking file existence, and getting file paths.

23. What are the challenges associated with memory management in Python?
 - Memory leaks due to circular references
 - High memory usage in large data processing
 - Garbage collection pauses affecting performance
 - Difficulty in controlling exact memory release timing

24. How do you raise an exception manually in Python?
We use the raise keyword.
Example: raise ValueError("Invalid input")

25. Why is it important to use multithreading in certain applications?
Multithreading is important for I/O-bound applications (like web scraping, network requests, file operations) because it allows the program to contiue working while waiting for external resources, improving responsiveness and efficiency.

````python
# 1. How can you open a file for writing in Python and write a string to it?
with open("output.txt", "w") as f:
  f.write("Hello, this is a sample text.")
print("Data written successfully.")
````

````python
# 2. Write a Python program to read the contents of a file and print each line.
# Practical Question 2
# Program to read the contents of a file and print each line

# First, we create a sample file if it doesn't exist (for demonstration)
with open("sample.txt", "w") as f:
    f.write("Hello World\n")
    f.write("Welcome to Python Programming\n")
    f.write("This is line number 3\n")
    f.write("Python is easy to learn\n")

print("Sample file 'sample.txt' created successfully.\n")

# Now reading the file and printing each line
try:
    with open("sample.txt", "r") as f:
        print("Contents of the file:")
        print("-" * 40)
        for line in f:
            print(line.strip())   # strip() removes extra newline characters
        print("-" * 40)

except FileNotFoundError:
    print("Error: The file 'sample.txt' does not exist.")
except Exception as e:
    print("An unexpected error occurred:", e)
````

````python
# 6. Write a Python program that logs an error message to log file when a division by zero exception occurs.
import logging
logging.basicConfig(filename="error.log", level=logging.ERROR)

try:
  result = 10 / 0
except ZeroDivisionError as e:
  logging.error("Division by zero error occured: %s", e)
  print("Error logged.")
````

````python
# 18. How would you implement a basic logging setup that logs to a file with rotation after 1MB?
import logging
from logging.handlers import RotatingFileHandler

handler = RotatingFileHandler("app.log", maxBytes=1024*1024, backupCount=5)
logging.basicConfig(handlers=[handler], level=logging.INFO)
logging.info("This is a test log message.")
````

````python
# 19. Write a program that handles both IndexError and KeyError using a try-except block.
my_list = [1, 2, 3]
my_dict = {"a": 1}

try:
    print(my_list[5])      # IndexError
    print(my_dict["b"])    # KeyError
except IndexError:
    print("Index out of range.")
except KeyError:
    print("Key not found in dictionary.")
````

````python
# 21. Write a Python program that reads a file and prints the number of occurrences of a specific word.
word = "python"
count = 0
with open("sample.txt", "r") as f:
    for line in f:
        count += line.lower().count(word.lower())
print(f"The word '{word}' appears {count} times.")
````

````python
# 22. How can you check if a file is empty before attempting to read its contents?
import os
if os.path.exists("sample.txt") and os.path.getsize("sample.txt") > 0:
    with open("sample.txt", "r") as f:
        print(f.read())
else:
    print("File is empty or does not exist.")
````
